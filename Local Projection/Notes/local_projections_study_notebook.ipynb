# Local Projections (LP) 

## Jorda (2005)
### Core intuition
* The key idea: LP estimates impulse responses by directly regressing future outcomes $y_{t+h}$ on current shocks, ​instead of deriving them from a full system like VAR.

* Advantages of estimate a separate regression for each horizon: allow the effect of a shock vary over time; avoids imposing potentially incorrect dynamic restrictions. Estimating a separate regression for each horizon allows the effect of a shock to vary flexibly over time and avoids imposing potentially incorrect dynamic restrictions across horizons.

* Definition of impulse response: the difference between forecasts

    $IR(t,s,d_i) = E(y_{t+x}|v_t = d_i; X_t) - E(y_{t+s}|v_t = 0; X_t)$
    
    which measures how a variable changes over time after a shock. It compares what happens if a shock occurs versus if it does not.

* Impulse resposne can be defined without specifying the full DGP: impulse responses depend only on the conditional expectation of future outcomes given current information and a shock. As long as condition on past information and identify the shock, its effect can be traced on future variables directly. Therefore, specifying the entire DGP is unnecessary and may even be harmful, since misspecification of the model can introduce forecast errors.

### VAR Limitation
* difference between VAR-based and LP-based impulse response: 
    VAR-based impulse responses rely on correctly specifying the full data-generating process, since the impulse responses are derived from the estimated VAR system. In contrast, local projection impulse responses are obtained directly by regressing future outcomes on current shocks and controls, without requiring full system specification.

    Additionally, VAR-based methods require identification assumptions, such as variable ordering (e.g., Cholesky decomposition), to recover structural shocks. If these assumptions are incorrect, the estimated shocks may reflect forecast errors rather than true structural shocks.

    Local projections avoid some of these issues by estimating impulse responses directly at each horizon

 
VAR requires a *two-step procedure*-- we first estimate the VAR system and then compute impulse responses by transforming it into its moving-average representation. This can be problematic because impulse responses are not estimated directly; instead, they depend on the estimated VAR coefficients. If the VAR is misspecified, these errors carry over and may be amplified when computing impulse responses, particularly at longer horizons.


### Regression Specification

General form of LP regression: 
$y_{t+s} = \alpha^{s} + B_1^{s+1} y_{t-1} + B_2^{s+1} y_{t-2} + \cdots + B_p^{s+1} y_{t-p} + u_{t+s}
\quad \text{for } s = 0, 1, 2, \ldots, h$

* it estimates each horizon independently rather than constructing forecasts recursively

* LP avoids the accumulation of errors that arises in VARs, since each horizon is estimated separately and does not depend on previous forecasts.

* allows for greater flexibility, as the underlying dynamics can differ across horizons.

* Because the residuals from these regressions incorporate multiple past shocks, they follow a moving-average structure, which induces serial correlation. For this reason, heteroskedasticity and autocorrelation consistent (HAC) standard errors are required for valid inference.
